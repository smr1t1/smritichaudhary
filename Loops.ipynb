```python
list_of_numbers = [1,2,4,6,11,14,17,20]
for number in list_of_numbers:
  squared_number = number**2
  print(f"The square of {number} is {squared_number}")
```

```python
list_of_numbers
```

```python
squared_numbers = []

for number in list_of_numbers:
  square = number**2
  squared_numbers.append(square)
print(f"The list of squared numbers is {squared_numbers}")
```

```python
print(list_of_numbers)
print(squared_numbers)
```

```python
zipped_dict = dict(zip(list_of_numbers, squared_numbers))
print(zipped_dict)
```

```python
for numbers in list_of_numbers:
  if number%2 == 0:
    squared_numbers = number**2
    print(f"The square of {number} is {squared_numbers}")
  else:
    print(f"I am an odd number {number}. My master prohibits me from printing the square of odd numbers. Strange but okay!")
```

```python
heisenberg_quote = "I am the one who knocks!"
for char in heisenberg_quote:
  print(f" The character is {char}")
```

```python
country_city_river_list = [('India','New Delhi','Ganga',),('Australia','Canberra','Rovers'),('United States','Wahington DC','Missouri'),('England','London','Thames')]
for country,city,river in country_city_river_list:
  print(f"The capital of {country} is {city} and it also has the river {river}.")
```

```python
country_city_dict = dict(country_city_list)
print(country_city_dict)
```

```python
country_city_dict = {1:"India",2:"USA"}
for item in country_city_dict:
  print(item,country_city_dict[item])
```

```python
x= 0
while x < 10:
  print('x is currently: ',x)
  print('x is still less than 10, adding 1 to x')
  x= x+1
```
